```python
sequence_lengths = ['10 à 20', '21 à 30', '31 à 40', '41 à 50']

transformer_mfe = [0.03, 0.095, 0.231, 0.503]

lstm_mfe = [0.011, 0.073, 0.238, 0.457]

cnn_mfe = [0.105, 0.166, 0.323, 0.561]



# Plotting

plt.figure(figsize=(10, 6))



plt.plot(sequence_lengths, transformer_mfe, marker='o', label='Transformer')

plt.plot(sequence_lengths, lstm_mfe, marker='s', label='LSTM')

plt.plot(sequence_lengths, cnn_mfe, marker='^', label='CNN')



plt.xlabel('Longueur de séquence')

plt.ylabel('MSE')

plt.title('Perte MFE vs. Longueur de séquences de chaque modèle')

plt.legend()



# Show the plot

plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table provided
sequence_intervals = ['10 to 20', '21 to 30', '31 to 40', '41 to 50']

# Transformer data
transformer_mfe = np.array([0.01707, 0.00421, 0.00649, 0.01835])
transformer_hairpins = np.array([0.0004, 0.00043, 0.00052, 0.00086])

# Transformer data
transformer_mfe_2 = np.array([0.03, 0.095, 0.231, 0.503])
transformer_hairpins_2 = np.array([0.007, 0.043, 0.110, 0.182])

# Calculate total loss for each model
transformer_total_loss = transformer_mfe + transformer_hairpins
transformer_total_loss_2 = transformer_mfe_2 + transformer_hairpins_2
#lstm_total_loss = lstm_mfe + lstm_hairpins
#cnn_total_loss = cnn_mfe + cnn_hairpins

# Plotting total loss for each model
plt.figure(figsize=(10, 5))
plt.bar(sequence_intervals, transformer_total_loss, color='skyblue', label='Perte totale Transformer*')
plt.bar(sequence_intervals, transformer_total_loss_2, color='orange', label='Perte totale Transformer')
#plt.plot(sequence_intervals, cnn_total_loss, '-^', label='CNN Total Loss')
plt.title('Perte totale à différentes longueurs de séquences')
plt.xlabel('Longueur des séquences')
plt.ylabel('Perte totale (MFE + hairpins)')
plt.legend()
plt.grid(True)
plt.show()

```
